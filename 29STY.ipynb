```python
%%file hello.c

int add(int a, int b){
    return a + b;
}
```

```python
!gcc -shared hello.c -o hello.so
```

```python
!file *
```

```python
import ctypes

lib = ctypes.CDLL("./hello.so")
add = lib.add

add.argtypes = ctypes.c_int, ctypes.c_int
add.restype = ctypes.c_int

print(add(0, 9))
```
